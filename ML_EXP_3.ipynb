```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('/content/User_Data.csv')
df.head()
x=df.iloc[:,[2,3]].values
y=df.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
sd= StandardScaler()
x_train=sd.fit_transform(x_train)
x_test=sd.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier(criterion="gini",random_state=100)
dc.fit(x_train,y_train)
```

```python
y_pred=dc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print ("Accuracy : ",accuracy_score(y_test,y_pred)*100)
print("Report : ",classification_report(y_test, y_pred))
```

```python
from sklearn import tree
tree.plot_tree(dc,feature_names=(y),class_names=None,filled=True)
fig = plt.figure(figsize=(25,20))
fig.savefig("decistion_tree.png")
```
